```python
import pandas as pd, numpy as np
pd.options.display.max_columns = 100
```

```python
median_tod_delays = pd.read_csv('delay idx tables/Large/median tod delays.csv', sep=';')
median_toy_delays = pd.read_csv('delay idx tables/Large/median toy delays.csv', sep=';')
mode_tod_delays = pd.read_csv('delay idx tables/Large/mode tod delays.csv', sep=';')
mode_toy_delays = pd.read_csv('delay idx tables/Large/mode toy delays.csv', sep=';')
flights = pd.read_csv('../Data Exploration/samples/flights_sample.csv', sep=';', low_memory=False)
flights.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
```

```python
print(len(median_tod_delays))
median_tod_delays.tail()
```

```python
car_dep_median_hourly_delays = median_tod_delays.groupby(by=['Scheduled Departure (Hour)', 'Airline']).median()
car_dep_median_hourly_delays.drop(columns='Scheduled Arrival (Hour)', inplace=True)
car_dep_median_hourly_delays.reset_index(inplace=True)
car_dep_median_hourly_delays
```

```python
car_arr_median_hourly_delays = median_tod_delays.groupby(by=['Scheduled Arrival (Hour)', 'Airline']).median()
car_arr_median_hourly_delays.drop(columns='Scheduled Departure (Hour)', inplace=True)
car_arr_median_hourly_delays.reset_index(inplace=True)
car_arr_median_hourly_delays
```

```python
median_toy_delays.head()
```

```python
car_mode_monthly_delays = mode_toy_delays.groupby(by=['Month', 'Airline']).agg(pd.Series.mode)
car_mode_monthly_delays.drop(columns=['Departing Airport', 'Arriving Airport'], inplace=True)
car_mode_monthly_delays.reset_index(inplace=True)
mode_list = []
for row in range(len(car_mode_monthly_delays)):
    if 'ndarray' in str(type(car_mode_monthly_delays['Monthly Arrival Delay(Mode)'][row])):
        mode_list.append(car_mode_monthly_delays['Monthly Arrival Delay(Mode)'][row].mean()) #if there are multiple modes, take the mean of them all

    else:
        mode_list.append(car_mode_monthly_delays['Monthly Arrival Delay(Mode)'][row])
        
car_mode_monthly_delays['Monthly Arrival Delay(Mode)'] = mode_list
car_mode_monthly_delays
```

```python
median_toy_delays.head()
```

```python
car_median_monthly_delays = median_toy_delays.groupby(by=['Month', 'Airline']).median()
car_median_monthly_delays.reset_index(inplace=True)
car_median_monthly_delays
```

```python
car_median_monthly_delays.to_csv('delay idx tables/car_median_monthly_delays.csv', sep=';')
car_mode_monthly_delays.to_csv('delay idx tables/car_mode_monthly_delays.csv', sep=';')
car_arr_median_hourly_delays.to_csv('delay idx tables/car_arr_median_hourly_delays.csv', sep=';')
car_dep_median_hourly_delays.to_csv('delay idx tables/car_dep_median_hourly_delays.csv', sep=';')
```
